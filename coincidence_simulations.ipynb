```python
%matplotlib inline             
import matplotlib.pyplot as plt
import numpy as np
import math

#Function definition

def coincidence(delta_i, delta_s, phi):
    return 1+math.cos(delta_i + delta_s + phi)
```

```python
#Value definitions 
t_1 = 0.5
t_2 = 0.3
t_3 = 0.2
t_4 = 0.1
#Case when both deltas change independently
w_i1 = np.full(1000,100)
w_s1 = np.linspace(100,350,1000)
w_i2 = np.full(1000, 120)
w_s2 = np.linspace(120,400, 1000)
values_c = np.array([coincidence((w_s1[i] + w_s2[i])*(t_1-t_3), (w_i1[i] + w_i2[i])*(t_2-t_4), w_i1[i]*t_4 - w_i2[i]*t_2 + w_s1[i]*t_3 - w_s2[i]*t_1) for i in range(0,300)])

#Case with deltas changing equally, wp remains always constant
w_p = 300
w_s1 = np.linspace(100,300,1000)
w_i1 = np.array([w_p - w_s1[i] for i in range(0,1000)])
w_s2 = np.linspace(100,300,1000)
w_i2 = np.array([w_p - w_s2[i] for i in range(0,1000)])
values_c2 = np.array([coincidence((w_s1[i] + w_s2[i])*(t_1-t_3), (w_i1[i] + w_i2[i])*(t_2-t_4), w_i1[i]*t_4 - w_i2[i]*t_2 + w_s1[i]*t_3 - w_s2[i]*t_1) for i in range(0,300)])

#Case when only one of the deltas change
w_p = 300
w_s1 = np.linspace(100,300,1000)
w_i1 = np.array([w_p - w_s1[i] for i in range(0,1000)])
w_i2 = np.full(1000, 120)
w_s2 = np.full(1000,180)
values_c3 = np.array([coincidence((w_s1[i] + w_s2[i])*(t_1-t_3), (w_i1[i] + w_i2[i])*(t_2-t_4), w_i1[i]*t_4 - w_i2[i]*t_2 + w_s1[i]*t_3 - w_s2[i]*t_1) for i in range(0,300)])


plt.plot(np.linspace(0,500,300),values_c, label="Both deltas change independently")
plt.plot(np.linspace(0,500,300),values_c2, label="Both Deltas change equally")
plt.plot(np.linspace(0,500,300),values_c3, label="Only delta_i changes")
plt.ylabel("Coincidence Count Rate")
plt.xlabel("delta_i + delta_s + phi")
plt.title("Coincidence")
plt.legend(loc="lower right")
plt.show()




```
